## Independent ttest

```python
import pandas as pd
from scipy.stats import ttest_ind, f_oneway
import scipy.stats as ST3

# 1. Independent t-test: Calories_Burned by Gender

# Separate groups
male_calories = al[al['Gender'] == 'Male']['Burned']
female_calories = al[al['Gender'] == 'Female']['Burned']

```

```python
# Perform t-test
t_stat, p_val = ST3.ttest_ind(male_calories, female_calories, equal_var=False)

print("=== Hypothesis Test 1: Calories Burned by Gender ===")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print(" Reject Null Hypothesis: There is a significant difference in calories burned between genders.")
else:
    print(" Fail to reject Null Hypothesis: No significant difference in calories burned between genders.")


```

## Chi_Square

```python

import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns



# Create a contingency table
contingency_table = pd.crosstab(al['Gender'], al['Workout'])

# Perform Chi-square test
chi2, p_val, dof, ex_frequency = chi2_contingency(contingency_table)

print("=== Hypothesis Chi Square test: Relationship between Gender and Workout Type ===")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print(" Reject H₀: There is a significant relationship between gender and workout type.")
else:
    print(" Fail to reject H₀: No significant relationship between gender and workout type.")
```
